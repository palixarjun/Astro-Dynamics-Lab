```python
# Initialize simulation
sim = rebound.Simulation()
sim.units = ('AU', 'yr', 'Msun')

# Add Sun
sim.add(m=1.0)

# Add Jupiter
sim.add(m=9.54e-4, a=5.2, e=0.048, inc=0.0)

# Add comet on a highly eccentric orbit crossing Jupiter's path
sim.add(m=0.0, a=6.0, e=0.8, inc=np.radians(5), omega=np.radians(0), Omega=np.radians(0), f=np.radians(0))

# Use IAS15 for high-accuracy during close encounters
sim.integrator = "ias15"

# Integrate and print comet distance to Jupiter over time
times = np.linspace(0, 5, 10)  # 5 years
for t in times:
    sim.integrate(t)
    comet = sim.particles[2]
    jupiter = sim.particles[1]
    r = np.sqrt((comet.x-jupiter.x)**2 + (comet.y-jupiter.y)**2 + (comet.z-jupiter.z)**2)
    print(f"Time = {t:.2f} yr, Comet-Jupiter distance = {r:.3f} AU")
```

```python
# Initialize Simulation Function
def init_solar_system_comet():

    # Initialize Sun + Jupiter + comet system for integrator comparison.
    # Returns a REBOUND simulation object.

    sim = rebound.Simulation()
    sim.units = ('AU', 'yr', 'Msun')
    
    # Sun
    sim.add(m=1.0)
    
    # Jupiter
    sim.add(m=9.54e-4, a=5.2, e=0.048, inc=0.0)
    
    # Comet on a highly eccentric orbit crossing Jupiter's path
    sim.add(m=0.0, a=6.0, e=0.8, inc=np.radians(5), omega=0, Omega=0, f=0)
    
    return sim
```

```python
# Function: Integrate and record comet positions
def integrate_positions(sim, integrator, times):

    # Integrate simulation using the specified integrator.
    
    # Returns: positions: Nx3 array of comet (x,y,z) positions at each time step.
    sim_copy = sim.copy()
    sim_copy.integrator = integrator
    if integrator=="whfast":
        sim_copy.dt = 0.001  # small timestep for WHFast
    
    positions = []
    for t in times:
        sim_copy.integrate(t)
        comet = sim_copy.particles[2]
        positions.append([comet.x, comet.y, comet.z])
    
    return np.array(positions)
```

```python
# Main Program
# Time array for integration
total_years = 5
n_steps = 500
times = np.linspace(0, total_years, n_steps)

# Initialize system
sim = init_solar_system_comet()

# Integrate using IAS15 (reference, high accuracy)
ias15_positions = integrate_positions(sim, "ias15", times)

# Integrate using WHFast
whfast_positions = integrate_positions(sim, "whfast", times)

# Compute Error Magnitudes
errors = np.linalg.norm(ias15_positions - whfast_positions, axis=1)

# Print sample errors every 50 steps
print("Sample error magnitudes between IAS15 and WHFast:")
for i in range(0, n_steps, 50):
    print(f"Time = {times[i]:.2f} yr, Error = {errors[i]:.6f} AU")

# Plot Error Over Time
plt.figure(figsize=(8,4))
plt.plot(times, errors, color='purple', lw=2)
plt.xlabel("Time [yr]")
plt.ylabel("Position Error Magnitude [AU]")
plt.title("Comet Orbit Error: WHFast vs IAS15")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Function: Run simulation for a given integrator
def run_sim(integrator_name, dt=None):
    sim = rebound.Simulation()
    sim.units = ('AU', 'yr', 'Msun')
    
    # Sun
    sim.add(m=1.0)
    
    # Jupiter
    sim.add(m=9.54e-4, a=5.2, e=0.048)
    
    # Asteroid crossing Jupiter orbit
    sim.add(m=0.0, a=5.2, e=0.6)
    
    sim.move_to_com()
    sim.integrator = integrator_name
    
    if dt:
        sim.dt = dt
    
    times = np.linspace(0, 50, 2000)  # 50 years
    
    a_vals = []
    E0 = sim.energy()
    dE = []
    
    for t in times:
        sim.integrate(t)
        orbit = sim.orbits()[1]   # asteroid
        a_vals.append(orbit.a)
        dE.append((sim.energy()-E0)/E0)
    
    return times, np.array(a_vals), np.array(dE)
```

```python
# Run IAS15 (reference)
t_ias, a_ias, dE_ias = run_sim("ias15")

# Run WHFast
t_wh, a_wh, dE_wh = run_sim("whfast", dt=0.001)
```

```python
print("IAS15 max |dE|:", np.max(np.abs(dE_ias)))
print("WHFast max |dE|:", np.max(np.abs(dE_wh)))

plt.figure(figsize=(16,6))  # wider figure for elongation

# Plot with slightly thicker lines and alpha for clarity
plt.plot(t_ias, np.abs(dE_ias), color='blue', lw=2, linestyle='-', alpha=0.8, label='IAS15')
plt.plot(t_wh, np.abs(dE_wh), color='gold', lw=2, linestyle='-', alpha=0.8, label='WHFast')

plt.yscale('log')  # logarithmic y-axis

plt.xlabel("Time [years]", fontsize=12)
plt.ylabel("|ΔE/E0| (log scale)", fontsize=12)
plt.title("IAS15 vs WHFast: Energy Conservation Error (Solid Lines)", fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()  # avoids clipping of labels
plt.show()
```
