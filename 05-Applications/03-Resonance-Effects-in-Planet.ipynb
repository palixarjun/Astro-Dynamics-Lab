```python
# Planetary Resonance Effect

# 1. Definition:
#    Planetary resonance occurs when two orbiting bodies (planets, moons, asteroids) 
#    exert regular, periodic gravitational influence on each other because their 
#    orbital periods are in a ratio of small integers (e.g., 2:1, 3:2).
#
# 2. Examples:
#    - Jupiter's moons Io, Europa, and Ganymede are in a 4:2:1 Laplace resonance.
#    - Pluto and Neptune are in a 3:2 mean-motion resonance.
#
# 3. Effects of Resonance:
#    - Can stabilize orbits (prevent close encounters or collisions)
#    - Can increase eccentricity or inclination over time
#    - Responsible for gaps in asteroid belts (Kirkwood gaps)
#
# 4. Mean-Motion Resonance (MMR):
#    - If planet A completes p orbits while planet B completes q orbits, the system
#      is in a p:q mean-motion resonance.
#
# 5. Importance:
#    - Explains orbital configurations in exoplanet systems
#    - Helps understand planetary migration and long-term stability
#    - Key in predicting resonant chains in tightly packed systems
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Visualizing a Simple 2:1 Resonance
# Orbital periods (years)
T_inner = 1.0       # Inner planet
T_outer = 2.0       # Outer planet (2:1 resonance)

# Time array
t = np.linspace(0, 4, 1000)  # simulate 4 years

# Angular positions (theta = 2*pi*t / T)
theta_inner = 2*np.pi * t / T_inner
theta_outer = 2*np.pi * t / T_outer

# Orbital radii (AU)
r_inner = 1.0
r_outer = 1.6

# Cartesian positions
x_inner = r_inner * np.cos(theta_inner)
y_inner = r_inner * np.sin(theta_inner)

x_outer = r_outer * np.cos(theta_outer)
y_outer = r_outer * np.sin(theta_outer)

# Plot orbits
plt.figure(figsize=(6,6))
plt.plot(0,0,'*', color='orange', markersize=15, label='Star')
plt.plot(x_inner, y_inner, 'b', label='Inner Planet (1 yr)')
plt.plot(x_outer, y_outer, 'g', label='Outer Planet (2 yr)')
plt.xlabel("X (AU)")
plt.ylabel("Y (AU)")
plt.title("2:1 Orbital Resonance")
plt.axis('equal')
plt.legend()
plt.show()
```

```python
# Inner planet: completes 1 orbit per year.
# Outer planet: completes 1 orbit every 2 years → exactly 2:1 mean-motion resonance.
# Over 4 years, you can see the periodic alignment every 2 years.
# This is the simplest visual demonstration of resonance.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Orbital radii (AU)
r_inner = 1.0
r_outer = 1.5874  # Choose so that outer completes half the angular speed (2:1 resonance)

# Orbital periods (years)
T_inner = 1.0
T_outer = 2.0

# Time array
t = np.linspace(0, 4, 500)  # simulate 4 years

# Angular positions
theta_inner = 2*np.pi * t / T_inner
theta_outer = 2*np.pi * t / T_outer

# Cartesian coordinates
x_inner = r_inner * np.cos(theta_inner)
y_inner = r_inner * np.sin(theta_inner)
x_outer = r_outer * np.cos(theta_outer)
y_outer = r_outer * np.sin(theta_outer)

# Plot
plt.figure(figsize=(4,4))
plt.plot(0, 0, '*', color='orange', markersize=15, label='Star')
plt.plot(x_inner, y_inner, 'b-', alpha=0.7, label='Inner Planet Orbit')
plt.plot(x_outer, y_outer, 'g-', alpha=0.7, label='Outer Planet Orbit')

# Mark positions at integer years
for year in range(5):
    idx = np.argmin(np.abs(t - year))
    plt.plot(x_inner[idx], y_inner[idx], 'bo', markersize=8)
    plt.plot(x_outer[idx], y_outer[idx], 'go', markersize=8)
    plt.text(x_outer[idx]+0.05, y_outer[idx]+0.05, f"t={year}yr", fontsize=9)

plt.xlabel("X (AU)")
plt.ylabel("Y (AU)")
plt.title("2:1 Orbital Resonance: 2D Trajectories")
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.legend(loc='upper right')
plt.show()
```

```python
# Outer planet radius adjusted (r_outer = 1.5874 AU) so that angular speeds match the 2:1 period ratio.
# Positions at integer years now align perfectly with inner planet, demonstrating the resonance clearly.
# The green orbit now correctly shows periodicity without distortion.
```

```python
# Inner planet
a_inner = 1.0
T_inner = 1.0

# Resonance ratios (p:q)
resonances = {
    "2:1": (2,1),
    "3:2": (3,2),
    "3:1": (3,1)
}

t = np.linspace(0, 3, 1000)  # simulate 3 years

fig, axes = plt.subplots(1, 3, figsize=(18,6))

for ax, (label, (p,q)) in zip(axes, resonances.items()):
    
    T_outer = p/q
    a_outer = T_outer**(2/3)  # from Kepler's 3rd law
    
    theta_inner = 2*np.pi * t / T_inner
    theta_outer = 2*np.pi * t / T_outer
    
    x_inner = a_inner * np.cos(theta_inner)
    y_inner = a_inner * np.sin(theta_inner)
    
    x_outer = a_outer * np.cos(theta_outer)
    y_outer = a_outer * np.sin(theta_outer)
    
    ax.plot(0,0,'*',color='orange',markersize=15)
    ax.plot(x_inner, y_inner, 'b-', alpha=0.7)
    ax.plot(x_outer, y_outer, 'g-', alpha=0.7)
    
    ax.set_title(f"{label} Resonance")
    ax.set_aspect('equal')
    ax.set_xlim(-2.5,2.5)
    ax.set_ylim(-2.5,2.5)
    ax.grid(True)

plt.suptitle("Comparison of Orbital Resonances (2D View)", fontsize=16)
plt.show()
```

```python
# Orbital Resonance Comparison

# 2:1 Resonance
# - Outer planet completes 1 orbit for every 2 inner orbits.
# - Produces clear periodic alignment.
# - Strong gravitational locking is possible.
# - Often dynamically stable if spacing is sufficient.

# 3:2 Resonance
# - Outer planet completes 2 orbits for every 3 inner orbits.
# - Alignment pattern is more complex than 2:1.
# - Common in tightly packed exoplanet systems.
# - Can maintain long-term stability through resonance protection.

# 3:1 Resonance
# - Outer planet moves significantly slower.
# - Requires larger orbital spacing.
# - Can strongly excite orbital eccentricity.
# - Often associated with instability zones (e.g., asteroid gaps).
```

```python
# Computing Resonant Orbital Radii using Kepler's Law

# Assume star mass = 1 (solar mass units)
# Kepler's Third Law (for circular orbits in AU & years):
# P^2 = a^3  →  a = P^(2/3)

def semi_major_axis_from_period(P):
    return P**(2/3)

# Inner planet reference
P_inner = 1.0       # 1 year
a_inner = semi_major_axis_from_period(P_inner)

# Resonances
resonances = {
    "2:1": 2/1,
    "3:2": 3/2,
    "3:1": 3/1
}

print("Inner planet:")
print("Period =", P_inner, "years")
print("Semi-major axis =", round(a_inner,4), "AU\n")

print("Resonant outer planets:")

for name, ratio in resonances.items():
    P_outer = ratio * P_inner
    a_outer = semi_major_axis_from_period(P_outer)
    print(f"{name} resonance → Period = {P_outer:.2f} yr | a = {a_outer:.4f} AU")
```

```python
# Cell 2: 2D Visualization of Resonant Orbits
# Kepler function 
def semi_major_axis_from_period(P):
    return P**(2/3)

P_inner = 1.0
a_inner = semi_major_axis_from_period(P_inner)

resonances = {
    "2:1": 2/1,
    "3:2": 3/2,
    "3:1": 3/1
}

theta = np.linspace(0, 2*np.pi, 1000)

plt.figure(figsize=(7,7))

# Plot inner orbit
x_inner = a_inner * np.cos(theta)
y_inner = a_inner * np.sin(theta)
plt.plot(x_inner, y_inner, label="Inner Planet")

# Plot resonant outer orbits
for name, ratio in resonances.items():
    P_outer = ratio * P_inner
    a_outer = semi_major_axis_from_period(P_outer)
    
    x_outer = a_outer * np.cos(theta)
    y_outer = a_outer * np.sin(theta)
    
    plt.plot(x_outer, y_outer, label=f"{name} Resonance")

# Star at center
plt.scatter(0,0,s=150)

plt.gca().set_aspect('equal')
plt.xlabel("X (AU)")
plt.ylabel("Y (AU)")
plt.title("Orbital Spacing for Different Resonances")
plt.legend()
plt.grid(True)
plt.legend(loc = "upper right")
plt.show()
```

```python
# resonance alignment behavior (phase locking).
# Phase Alignment in 2:1 and 3:2 Resonance

t = np.linspace(0, 20, 8000)

P_inner = 1.0
omega_inner = 2*np.pi / P_inner

resonances = {
    "2:1": 2/1,
    "3:2": 3/2
}

plt.figure(figsize=(8,8))

for name, ratio in resonances.items():
    
    P_outer = ratio * P_inner
    omega_outer = 2*np.pi / P_outer
    
    a_inner = P_inner**(2/3)
    a_outer = P_outer**(2/3)
    
    x_inner = a_inner * np.cos(omega_inner * t)
    y_inner = a_inner * np.sin(omega_inner * t)
    
    x_outer = a_outer * np.cos(omega_outer * t)
    y_outer = a_outer * np.sin(omega_outer * t)
    
    # Improved alignment detection
    phase_diff = (omega_inner*t - omega_outer*t)
    align = np.where(np.abs(np.sin(phase_diff)) < 0.01)[0]
    
    plt.scatter(x_inner[align], y_inner[align], s=8, label=f"{name} alignment")

# Star
plt.scatter(0,0,s=250)

# Force equal scaling and fixed limits
max_radius = max([ (ratio*P_inner)**(2/3) for ratio in resonances.values() ])
plt.xlim(-max_radius*1.2, max_radius*1.2)
plt.ylim(-max_radius*1.2, max_radius*1.2)

plt.gca().set_aspect('equal')
plt.xlabel("X (AU)")
plt.ylabel("Y (AU)")
plt.title("Resonant Alignment Points (Clear View)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cell 4: Resonance Overlap Criterion (Wisdom 1980)

# Planet-to-star mass ratio
mu_values = np.logspace(-8, -2, 200)

# Wisdom criterion
delta_over_a = 1.3 * mu_values**(2/7)

plt.figure(figsize=(7,5))
plt.loglog(mu_values, delta_over_a)

plt.xlabel("Planet-Star Mass Ratio (μ)")
plt.ylabel("Minimum Stable Spacing Δa/a")
plt.title("Resonance Overlap Stability Limit")
plt.grid(True, which="both")
plt.show()
```

```python
# As planet mass increases:

# Resonances widen
# Overlap more easily
# Required spacing increases

# Small planets → can pack tightly
# Massive planets → must be widely spaced

# This is the physical limit behind:

# Compact exoplanet systems
# Jupiter clearing asteroid gaps
# Planet packing boundaries
```

```python
# Resonance Location Visualization

# Inner planet reference
a_inner = 1.0
P_inner = 1.0

def semi_major_axis_from_period(P):
    return P**(2/3)

# Resonance ratios
resonances = {
    "3:2": 3/2,
    "2:1": 2/1,
    "3:1": 3/1
}

# Compute resonance locations
a_res = {}

for name, ratio in resonances.items():
    P_outer = ratio * P_inner
    a_outer = semi_major_axis_from_period(P_outer)
    a_res[name] = a_outer

# Plot
plt.figure(figsize=(9,5))

# Baseline axis
plt.axhline(0)

# Mark inner planet
plt.scatter(a_inner, 0, s=150)
plt.text(a_inner, 0.02, "Inner Planet (1 AU)", ha='center')

# Mark resonances
for name, a_val in a_res.items():
    plt.axvline(a_val, linestyle='--')
    plt.text(a_val, 0.02, name, ha='center')

plt.xlim(0.8, 2.5)
plt.yticks([])
plt.xlabel("Semi-Major Axis (AU)")
plt.title("Locations of 3:2, 2:1, and 3:1 Resonances")
plt.show()
```

```python
# Applications of Orbital Resonance

# 1. Asteroid Belt Structure
# - Mean-motion resonances with Jupiter create Kirkwood gaps.
# - Resonant forcing increases eccentricity.
# - Unstable asteroids are removed from specific orbital zones.

# 2. Compact Exoplanet Systems
# - Many Kepler-discovered systems show 2:1 and 3:2 resonances.
# - Resonance allows tight packing while preserving stability.
# - Indicates past planetary migration.

# 3. Satellite Systems
# - Io, Europa, and Ganymede form a Laplace resonance.
# - Maintains tidal heating.
# - Influences internal geology and possible subsurface oceans.

# 4. Planet Migration
# - Migrating planets can become trapped in resonance.
# - Resonant chains reveal early disk evolution.

# 5. Planet Packing Limit
# - Resonance overlap sets minimum stable spacing.
# - Small planets → tighter packing possible.
# - Massive planets → require larger separations.
```
