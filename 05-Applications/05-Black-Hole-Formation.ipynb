```python
# Black Hole Formation — Introduction

# A black hole forms when a massive object's gravitational collapse
# becomes so extreme that not even light can escape.

# Stellar-mass black holes form from the core collapse of
# massive stars after nuclear fuel is exhausted.

# When internal pressure (from fusion) can no longer
# counteract gravity, the core collapses.

# If the remaining core mass exceeds the neutron
# degeneracy pressure limit (~3 solar masses),
# collapse continues toward a black hole.

# The boundary beyond which escape is impossible
# is called the event horizon.

# Black hole formation connects:
# - Stellar evolution
# - General relativity
# - Gravitational collapse physics
# - High-energy astrophysics
```

```python
import matplotlib.pyplot as plt
import numpy as np
import rebound
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
```

```python
# Gravitational Force Law
# Gravitational constant (SI units)
G = 6.67430e-11   # m^3 kg^-1 s^-2

# Example masses
M = 2.0e30       # mass of a star (kg)
m = 1.0          # test particle (kg)

# Distance from center (meters)
r = 7.0e8

# Newton's Law of Gravitation
F = G * M * m / r**2

print("Gravitational Force:", F, "Newtons")
```

```python
# Escape Velocity
G = 6.67430e-11   # gravitational constant

# Example: Sun-like mass
M = 2.0e30       # kg
r = 7.0e8        # meters (approx solar radius)

# Escape velocity formula
# v_escape = sqrt(2GM / r)

v_escape = np.sqrt(2 * G * M / r)

print("Escape velocity:", v_escape, "m/s")
```

```python
# Schwarzschild Radius
G = 6.67430e-11        # gravitational constant
c = 3.0e8             # speed of light (m/s)

# Example mass (Sun-like star)
M = 2.0e30            # kg

# Schwarzschild radius formula
# r_s = 2GM / c^2

r_s = 2 * G * M / c**2

print("Schwarzschild Radius:", r_s, "meters")
print("In kilometers:", r_s/1000, "km")
```

```python
# Black Hole Formation — Foundational Concepts

# 1. Newtonian Gravity
# - Gravitational force: F = GMm / r^2
# - Gravity strengthens as mass increases or radius decreases.
# - Extreme compression increases gravitational intensity.

# 2. Escape Velocity
# - Escape velocity: v_escape = sqrt(2GM / r)
# - Minimum speed required to escape gravity.
# - If escape velocity increases beyond physical limits,
#   objects cannot escape the gravitational field.

# 3. Schwarzschild Radius
# - r_s = 2GM / c^2
# - Radius where escape velocity equals speed of light.
# - If an object's physical radius becomes smaller than r_s,
#   an event horizon forms.
# - This condition defines a black hole.

# Logical Progression:
# Strong gravity → Increasing escape velocity →
# Escape velocity equals light speed →
# Event horizon formation → Black hole.
```

```python
# Gravitational Collapse Condition
G = 6.67430e-11

# Stellar core mass (kg)
M_core = 5.0e30      # larger than Sun

# Core radius (meters)
R_core = 1.0e4       # compressed core

# Approximate gravitational pressure scaling
# P_grav ~ GM^2 / R^4

P_grav = G * M_core**2 / R_core**4

print("Approximate Gravitational Pressure:", P_grav)
```

```python
# Gravitational Collapse Condition
G = 6.67430e-11

# Stellar core mass (kg)
M_core = 5.0e30      # larger than Sun

# Core radius (meters)
R_core = 1.0e4       # compressed core

# Approximate gravitational pressure scaling
# P_grav ~ GM^2 / R^4

P_grav = G * M_core**2 / R_core**4

print("Approximate Gravitational Pressure:", P_grav)
```

```python
# Stellar Collapse and Mass Limits

# 4. Gravitational Collapse Condition
# - Gravity pulls inward; internal pressure pushes outward.
# - Approximate gravitational pressure scales as:
#   P_grav ~ GM^2 / R^4
# - When fusion stops, thermal pressure vanishes.
# - Collapse proceeds if degeneracy pressure cannot balance gravity.

# Pressure Support Mechanisms:
# - Electron degeneracy pressure → supports white dwarfs.
# - Neutron degeneracy pressure → supports neutron stars.

# 5. Chandrasekhar Limit
# - Maximum mass supported by electron degeneracy pressure.
# - M_ch ≈ 1.4 Solar Masses.
# - If core mass < 1.4 M☉ → White dwarf forms.
# - If core mass > 1.4 M☉ → Collapse continues.

# Collapse Outcomes:
# - 1.4 M☉ < Mass < ~3 M☉ → Neutron star.
# - Mass > ~3 M☉ → Collapse unstoppable → Black hole formation.

# Logical Flow:
# Fusion stops → Pressure drops →
# Degeneracy pressure limit exceeded →
# Runaway gravitational collapse.
```

```python
# Tolman–Oppenheimer–Volkoff (TOV) Limit (Neutron Star Maximum Mass)
# Approximate TOV limit (in solar masses)
M_TOV_min = 2.0
M_TOV_max = 3.0

print("Tolman–Oppenheimer–Volkoff (TOV) Limit:")
print("~", M_TOV_min, "to", M_TOV_max, "Solar Masses")
```

```python
# Schwarzschild Metric (Conceptual Representation)
# Schwarzschild metric (outside a non-rotating spherical mass):

# ds^2 = -(1 - 2GM/(rc^2)) c^2 dt^2
#        + (1 - 2GM/(rc^2))^(-1) dr^2
#        + r^2 dΩ^2
```

```python
# Diagram of Schwarzschild Geometry
# Constants
G = 6.67430e-11
c = 3.0e8
M = 2.0e30   # Sun-mass black hole example

# Schwarzschild radius
r_s = 2 * G * M / c**2

# Radial coordinate (avoid r = r_s exactly)
r = np.linspace(r_s*1.05, r_s*10, 1000)

# Embedding surface equation (approximate)
z = 2 * np.sqrt(r_s * (r - r_s))

plt.figure(figsize=(5,4))
plt.plot(r/1000, z/1000)

plt.xlabel("Radial Distance (km)")
plt.ylabel("Embedding Depth (km)")
plt.title("Spacetime Curvature Around a Black Hole")
plt.grid(True)
plt.show()
```

```python
# Compare Spacetime Curvature for Different Masses
G = 6.67430e-11
c = 3.0e8

# Three masses
masses = {
    "1 Solar Mass": 2.0e30,
    "5 Solar Mass": 1.0e31,
    "20 Solar Mass": 4.0e31
}

plt.figure(figsize=(5,4))

for label, M in masses.items():
    
    r_s = 2 * G * M / c**2
    r = np.linspace(r_s*1.05, r_s*8, 800)
    z = 2 * np.sqrt(r_s * (r - r_s))
    
    plt.plot(r/1000, z/1000, label=label)

plt.xlabel("Radial Distance (km)")
plt.ylabel("Embedding Depth (km)")
plt.title("Spacetime Curvature vs Black Hole Mass")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# As mass increases:
# Schwarzschild radius increases
# Gravitational well becomes wider
# Curvature extends farther outward

# Important: 
# The shape is mathematically similar.
# Mass scales the system — it does not change the geometry type.
```

```python
# Gravitational Time Dilation (Schwarzschild Metric)
# Constants
G = 6.67430e-11
c = 3.0e8
M = 2.0e30   # 1 solar mass example

# Schwarzschild radius
r_s = 2 * G * M / c**2

# Radial distances (avoid exactly r_s)
r = np.linspace(r_s*1.01, r_s*10, 1000)

# Time dilation factor
# dτ/dt = sqrt(1 - r_s/r)
time_dilation = np.sqrt(1 - r_s/r)

plt.figure(figsize=(4,3))
plt.plot(r/1000, time_dilation)

plt.xlabel("Radial Distance (km)")
plt.ylabel("Proper Time / Far Observer Time")
plt.title("Gravitational Time Dilation Near a Black Hole")
plt.grid(True)
plt.show()
```

```python
# Gravitational Redshift Near a Black Hole
# Constants
G = 6.67430e-11
c = 3.0e8
M = 2.0e30

# Schwarzschild radius
r_s = 2 * G * M / c**2

# Radial distance (avoid r = r_s exactly)
r = np.linspace(r_s*1.01, r_s*10, 1000)

# Gravitational redshift
z = (1 / np.sqrt(1 - r_s/r)) - 1

plt.figure(figsize=(4,3))
plt.plot(r/1000, z)

plt.xlabel("Radial Distance (km)")
plt.ylabel("Gravitational Redshift (z)")
plt.title("Gravitational Redshift Near a Black Hole")
plt.grid(True)
plt.show()
```

```python
# Space and Time Together
# Constants
G = 6.67430e-11
c = 3.0e8
M = 2.0e30

r_s = 2 * G * M / c**2
r = np.linspace(r_s*1.05, r_s*10, 1000)

# Spatial curvature (embedding depth)
z = 2 * np.sqrt(r_s * (r - r_s))

# Time dilation factor
time_dilation = np.sqrt(1 - r_s/r)

fig, ax1 = plt.subplots(figsize=(4,3))

# Spatial curvature
ax1.plot(r/1000, z/1000)
ax1.set_xlabel("Radial Distance (km)")
ax1.set_ylabel("Spatial Curvature Depth (km)")
ax1.grid(True)

# Second axis for time
ax2 = ax1.twinx()
ax2.plot(r/1000, time_dilation)
ax2.set_ylabel("Proper Time / Far Time")

plt.title("Spacetime Around a Black Hole")
plt.show()
```

```python
# Event Horizon and Photon Sphere Visualization
# Constants
G = 6.67430e-11
c = 3.0e8
M = 2.0e30  # 1 solar mass

# Radii
r_s = 2 * G * M / c**2              # Event horizon
r_photon = 1.5 * r_s                # Photon sphere

# Create circle angles
theta = np.linspace(0, 2*np.pi, 1000)

# Event horizon circle
x_h = r_s * np.cos(theta)
y_h = r_s * np.sin(theta)

# Photon sphere circle
x_p = r_photon * np.cos(theta)
y_p = r_photon * np.sin(theta)

plt.figure(figsize=(4,4))
plt.plot(x_p/1000, y_p/1000, label="Photon Sphere (1.5 r_s)")
plt.plot(x_h/1000, y_h/1000, label="Event Horizon (r_s)")

plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel("km")
plt.ylabel("km")
plt.title("Event Horizon vs Photon Sphere")
plt.legend()
plt.grid(True)
plt.legend(loc = "upper right")
plt.show()
```

```python
# Clear Photon Bending Near Black Hole
# Constants
G = 6.67430e-11
c = 3.0e8
M = 2.0e30

r_s = 2 * G * M / c**2
r_ph = 1.5 * r_s

# X domain
x_vals = np.linspace(-10*r_s, 5*r_s, 4000)

plt.figure(figsize=(5,5))

# Different impact parameters
impact_parameters = [3.5*r_s, 2.5*r_s, 1.8*r_s, 1.6*r_s]

for b in impact_parameters:
    
    y_vals = []
    
    for x in x_vals:
        r = np.sqrt(x**2 + b**2)
        
        # Stronger bending model for visibility
        bend = (4*r_s / r)
        
        y = b - bend * (x / (abs(x)+1e-8))
        y_vals.append(y)
    
    plt.plot(x_vals/1000, np.array(y_vals)/1000)

# Event horizon
theta = np.linspace(0, 2*np.pi, 1000)
plt.fill((r_s*np.cos(theta))/1000,
         (r_s*np.sin(theta))/1000)

# Photon sphere
plt.plot((r_ph*np.cos(theta))/1000,
         (r_ph*np.sin(theta))/1000)

plt.gca().set_aspect('equal')
plt.xlim(-10*r_s/1000, 5*r_s/1000)
plt.ylim(-6*r_s/1000, 6*r_s/1000)

plt.title("Photon Trajectories Near Black Hole")
plt.xlabel("km")
plt.ylabel("km")
plt.grid(True)
plt.show()
```

```python
# Black Hole Shadow Silhouette with Legends
# Constants
G = 6.67430e-11
c = 3.0e8
M = 2.0e30

# Radii
r_s = 2 * G * M / c**2
r_ph = 1.5 * r_s
r_shadow = 2.6 * r_s

theta = np.linspace(0, 2*np.pi, 1000)

plt.figure(figsize=(5,5))

# Shadow (filled)
plt.fill(r_shadow*np.cos(theta)/1000,
         r_shadow*np.sin(theta)/1000,
         label="Observed Shadow (~2.6 r_s)")

# Photon sphere
plt.plot(r_ph*np.cos(theta)/1000,
         r_ph*np.sin(theta)/1000,
         linewidth=2,
         color = "orange",
         label="Photon Sphere (1.5 r_s)")

# Event horizon
plt.plot(r_s*np.cos(theta)/1000,
         r_s*np.sin(theta)/1000,
         linewidth=2,
         color = "yellow",
         label="Event Horizon (r_s)")

plt.gca().set_aspect('equal')

plt.xlabel("km")
plt.ylabel("km")
plt.title("Black Hole Shadow Structure")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Complete 2D Black Hole Structure Diagram
# Constants
G = 6.67430e-11
c = 3.0e8
M = 2.0e30

r_s = 2 * G * M / c**2
r_ph = 1.5 * r_s
r_shadow = 2.6 * r_s

theta = np.linspace(0, 2*np.pi, 1000)

plt.figure(figsize=(5,5))

# --- Shadow ---
plt.fill(r_shadow*np.cos(theta)/1000,
         r_shadow*np.sin(theta)/1000,
         alpha=0.3,
         label="Shadow (~2.6 r_s)")

# --- Photon Sphere ---
plt.plot(r_ph*np.cos(theta)/1000,
         r_ph*np.sin(theta)/1000,
         linewidth=2,
         label="Photon Sphere (1.5 r_s)")

# --- Event Horizon ---
plt.plot(r_s*np.cos(theta)/1000,
         r_s*np.sin(theta)/1000,
         linewidth=3,
         label="Event Horizon (r_s)")

# --- Accretion Disk (simple ring) ---
r_disk_inner = 3 * r_s
r_disk_outer = 5 * r_s

plt.plot(r_disk_inner*np.cos(theta)/1000,
         r_disk_inner*np.sin(theta)/1000,
         linestyle='--',
         label="Accretion Disk (inner)")

plt.plot(r_disk_outer*np.cos(theta)/1000,
         r_disk_outer*np.sin(theta)/1000,
         linestyle='--',
         label="Accretion Disk (outer)")

# --- Incoming Light Rays ---
x = np.linspace(-8*r_s, 0, 500)
for offset in [4*r_s, 3*r_s, 2*r_s]:
    y = offset - (4*r_s)/(np.sqrt(x**2 + offset**2))
    plt.plot(x/1000, y/1000)

plt.gca().set_aspect('equal')
plt.xlabel("km")
plt.ylabel("km")
plt.title("2D Structural Diagram of a Black Hole")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3D Black Hole
# Constants
G = 6.67430e-11
c = 3.0e8
M = 2.0e30

r_s = 2 * G * M / c**2
r_ph = 1.5 * r_s

fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(111, projection='3d')

# Mesh resolution (controls criss-cross density)
u = np.linspace(0, 2*np.pi, 40)
v = np.linspace(0, np.pi, 40)

# --- Event Horizon (Light Mesh) ---
x_h = r_s * np.outer(np.cos(u), np.sin(v))
y_h = r_s * np.outer(np.sin(u), np.sin(v))
z_h = r_s * np.outer(np.ones(np.size(u)), np.cos(v))

ax.plot_wireframe(x_h/1000, y_h/1000, z_h/1000, linewidth=0.6)

# --- Photon Sphere (Finer Mesh) ---
u2 = np.linspace(0, 2*np.pi, 30)
v2 = np.linspace(0, np.pi, 30)

x_p = r_ph * np.outer(np.cos(u2), np.sin(v2))
y_p = r_ph * np.outer(np.sin(u2), np.sin(v2))
z_p = r_ph * np.outer(np.ones(np.size(u2)), np.cos(v2))

ax.plot_wireframe(x_p/1000, y_p/1000, z_p/1000, linewidth=0.4)

# --- Accretion Disk (Tilted Ring) ---
theta = np.linspace(0, 2*np.pi, 500)
r_disk = 4 * r_s
tilt = np.radians(30)

x_d = r_disk * np.cos(theta)
y_d = r_disk * np.sin(theta)
z_d = np.zeros_like(theta)

y_d_tilt = y_d * np.cos(tilt)
z_d_tilt = y_d * np.sin(tilt)

ax.plot(x_d/1000, y_d_tilt/1000, z_d_tilt/1000, linewidth=2)

# ---- Equal scaling ----
max_range = 6 * r_s / 1000
ax.set_xlim(-max_range, max_range)
ax.set_ylim(-max_range, max_range)
ax.set_zlim(-max_range, max_range)
ax.set_box_aspect([1,1,1])

ax.set_xlabel("km")
ax.set_ylabel("km")
ax.set_zlabel("km")
ax.set_title("3D Schwarzschild Black Hole")

plt.show()
```

```python
# Constants
G = 6.67430e-11
c = 3.0e8
M = 2.0e30

r_s = 2 * G * M / c**2 / 1000   # Schwarzschild radius in km
r_ph = 1.5 * r_s                # Photon sphere radius in km

# Sphere grid
phi = np.linspace(0, 2*np.pi, 60)
theta = np.linspace(0, np.pi, 60)
phi, theta = np.meshgrid(phi, theta)

# Event Horizon
x_h = r_s * np.sin(theta) * np.cos(phi)
y_h = r_s * np.sin(theta) * np.sin(phi)
z_h = r_s * np.cos(theta)

# Photon Sphere
x_p = r_ph * np.sin(theta) * np.cos(phi)
y_p = r_ph * np.sin(theta) * np.sin(phi)
z_p = r_ph * np.cos(theta)

# Accretion Disk
theta_disk = np.linspace(0, 2*np.pi, 600)
r_disk = 4 * r_s
tilt = np.radians(30)  # tilt angle

# Circle in XY-plane
x_d = r_disk * np.cos(theta_disk)
y_d = r_disk * np.sin(theta_disk)
z_d = np.zeros_like(theta_disk)

# Rotate around X-axis for tilt
y_d_tilt = y_d * np.cos(tilt) - z_d * np.sin(tilt)
z_d_tilt = y_d * np.sin(tilt) + z_d * np.cos(tilt)

# Plot
fig = go.Figure()

# Event Horizon
fig.add_surface(x=x_h, y=y_h, z=z_h, opacity=0.6, showscale=False, name="Event Horizon")

# Photon Sphere
fig.add_surface(x=x_p, y=y_p, z=z_p, opacity=0.2, showscale=False, name="Photon Sphere")

# Accretion Disk
fig.add_trace(go.Scatter3d(x=x_d, y=y_d_tilt, z=z_d_tilt, mode='lines',
                           line=dict(color='orange', width=3), name="Accretion Disk"))

# Layout
fig.update_layout(
    title="Schwarzschild Black Hole",
    scene=dict(
        xaxis_title="km",
        yaxis_title="km",
        zaxis_title="km",
        aspectmode="cube"  # ensures correct proportions
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

fig.show()

# Output is not visible in github webpage, refer to Folder/Output-Plotly
```
