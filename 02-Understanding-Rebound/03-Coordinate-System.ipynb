```python
import numpy as np #import numpy library
import matplotlib.pyplot as plt #import matplotlib library
from mpl_toolkits.mplot3d import Axes3D #import a module from Axes3D
```

```python
# cartesian plane

# Define points
A = np.array([2, 3])
B = np.array([-1, 4])
C = np.array([3, -2])

# Plot
plt.figure()

plt.scatter(*A)
plt.scatter(*B)
plt.scatter(*C)

# Label points
plt.text(A[0], A[1], " A(2,3)")
plt.text(B[0], B[1], " B(-1,4)")
plt.text(C[0], C[1], " C(3,-2)")

# Axes lines
plt.axhline(0)
plt.axvline(0)

plt.gca().set_aspect('equal')
plt.grid()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Three Points in Cartesian Coordinate System")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Define 3D points
A = np.array([2, 3, 1])
B = np.array([-1, 4, 2])
C = np.array([3, -2, -1])

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter points
ax.scatter(*A)
ax.scatter(*B)
ax.scatter(*C)

# Label points
ax.text(*A, " A(2,3,1)")
ax.text(*B, " B(-1,4,2)")
ax.text(*C, " C(3,-2,-1)")

# Axes labels
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

ax.set_title("Three Points in 3D Cartesian Coordinate System")

plt.show()
```

```python
# adding vectors to this plane
# Compute vectors from A
AB = B - A
AC = C - A

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot points
ax.scatter(*A)
ax.scatter(*B)
ax.scatter(*C)

# Label points
ax.text(*A, " A")
ax.text(*B, " B")
ax.text(*C, " C")

# Plot vectors using quiver
ax.quiver(*A, *AB)
ax.quiver(*A, *AC)

# Axes labels
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("3D Points with Vectors")

plt.show()
```

```python
# Assume equal masses

r1 = A
r2 = B
r3 = C

# Jacobi 1
R1 = r1

# Jacobi 2
R2 = r2 - r1

# Center of mass of first two
r_cm = (r1 + r2) / 2

# Jacobi 3
R3 = r3 - r_cm

print("Jacobi Coordinates:")
print("R1 =", R1)
print("R2 =", R2)
print("R3 =", R3)
```

```python
# --- Jacobi Visualization ---

# Checkpoint 1
A = np.array([2, 3, 1])
B = np.array([-1, 4, 2])
C = np.array([3, -2, -1])

# Equal masses assumed
r_cm = (A + B) / 2

# Jacobi vectors
R2 = B - A
R3 = C - r_cm

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Original points
ax.scatter(*A)
ax.scatter(*B)
ax.scatter(*C)
ax.scatter(*r_cm)

ax.text(*A, " A")
ax.text(*B, " B")
ax.text(*C, " C")
ax.text(*r_cm, " CM")

# Jacobi vectors
ax.quiver(*A, *R2)      # R2
ax.quiver(*r_cm, *R3)   # R3

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Jacobi Coordinate Representation")

plt.show()
```

```python
# --- Barycentric Coordinates ---

# Compute global center of mass
CM_total = (A + B + C) / 3

# Shift coordinates
A_b = A - CM_total
B_b = B - CM_total
C_b = C - CM_total

print("Center of Mass:", CM_total)
print("Barycentric Coordinates:")
print("A' =", A_b)
print("B' =", B_b)
print("C' =", C_b)
```

```python
# Checkpoint 1
A = np.array([2, 3, 1])
B = np.array([-1, 4, 2])
C = np.array([3, -2, -1])

# Compute center of mass (equal masses)
CM = (A + B + C) / 3

# Barycentric coordinates
A_b = A - CM
B_b = B - CM
C_b = C - CM

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Original points
ax.scatter(*A)
ax.scatter(*B)
ax.scatter(*C)
ax.scatter(*CM)

ax.text(*A, " A")
ax.text(*B, " B")
ax.text(*C, " C")
ax.text(*CM, " CM")

# Vectors from CM to original points
ax.quiver(*CM, *(A - CM))
ax.quiver(*CM, *(B - CM))
ax.quiver(*CM, *(C - CM))

# Also plot shifted barycentric points around origin
ax.scatter(*A_b)
ax.scatter(*B_b)
ax.scatter(*C_b)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Barycentric Coordinate Shift (Origin → CM)")

plt.show()
```

```python
# numerical comparison
import numpy as np

# Checkpoint 1
A = np.array([2, 3, 1])
B = np.array([-1, 4, 2])
C = np.array([3, -2, -1])

# --- Cartesian ---
r1, r2, r3 = A, B, C

# --- Jacobi ---
R1 = r1
R2 = r2 - r1
CM12 = (r1 + r2) / 2
R3 = r3 - CM12

# --- Barycentric ---
CM = (r1 + r2 + r3) / 3
A_b = r1 - CM
B_b = r2 - CM
C_b = r3 - CM

print("CARTESIAN")
print(r1, r2, r3)

print("\nJACOBI")
print(R1, R2, R3)

print("\nBARYCENTRIC")
print(A_b, B_b, C_b)

print("\nCheck sum (barycentric):", A_b + B_b + C_b)
```

```python
# Graphical comparison (improved)

fig = plt.figure(figsize=(15,5))

# Common axis limits
limit = 6

# --- Cartesian ---
ax1 = fig.add_subplot(131, projection='3d')
ax1.scatter(*r1)
ax1.scatter(*r2)
ax1.scatter(*r3)
ax1.set_title("Cartesian")
ax1.set_xlim(-limit, limit)
ax1.set_ylim(-limit, limit)
ax1.set_zlim(-limit, limit)

# --- Jacobi ---
ax2 = fig.add_subplot(132, projection='3d')

# R1 at origin
ax2.scatter(0,0,0)

# R2 endpoint
R2_end = R2
ax2.scatter(*R2_end)
ax2.quiver(0,0,0,*R2, length=1, normalize=False)

# R3 endpoint
R3_end = CM12 + R3
ax2.scatter(*R3_end)
ax2.quiver(*CM12,*R3, length=1, normalize=False)

# Also show CM12
ax2.scatter(*CM12)

ax2.set_title("Jacobi")
ax2.set_xlim(-limit, limit)
ax2.set_ylim(-limit, limit)
ax2.set_zlim(-limit, limit)

# --- Barycentric ---
ax3 = fig.add_subplot(133, projection='3d')
ax3.scatter(*A_b)
ax3.scatter(*B_b)
ax3.scatter(*C_b)
ax3.set_title("Barycentric")
ax3.set_xlim(-limit, limit)
ax3.set_ylim(-limit, limit)
ax3.set_zlim(-limit, limit)

# Axis labels
for ax in [ax1, ax2, ax3]:
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")

plt.tight_layout()
plt.show()
```
