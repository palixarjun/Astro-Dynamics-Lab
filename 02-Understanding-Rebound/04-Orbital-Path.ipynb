```python
import rebound # importing the module - rebound

sim = rebound.Simulation() # create an n-body simulation

sim.add(m=1) # adds a star
sim.add(m=0.1, e=0.041, a=0.4, inc=0.2, f=0.43, Omega=0.82, omega=2.98) # body 2
sim.add(m=1e-3, e=0.24, a=1.0, pomega=2.14) # body 3
sim.add(m=1e-3, e=0.24, a=1.5, omega=1.14, l=2.1) # body 4
sim.add(a=-2.7, e=1.4, f=-1.5,omega=-0.7) # body 5, hyperbolic orbit

%matplotlib inline
op = rebound.OrbitPlot(sim, color = True, figsize=(4,4)) 
# computes the keplerian orbits
# plots them in 2D
# color each orbit differently
# shows elliptical and hyperbolic effect

# final summary
# creates a 5 body gravitational system
# contains 3 bound elliptical orbits
# contains one hyperbolic flyby
# plots the orbital shape around the center of mass
```

```python
%matplotlib inline
op = rebound.OrbitPlotSet(sim, color = True, figsize=(4,4))

# shows them in xy, yz and xz plane
```

```python
op = rebound.OrbitPlot(sim, # simulation object containing all bodies and orbital systems
unitlabel="[AU]", # labels the axes with astronimical units
color=True, # assign different colors to each orbit for visual distinction
periastron=True, # Marks the periastron point (closest approach to the central body) on each orbit.
xlim=[-2,2], # Sets horizontal axis limits from −2 to 2 AU.
ylim=[-2.5,1.5], # Sets horizontal axis limits from −2 to 2 AU.
figsize=(4,4)) # sets the figure size to 4x4 inches (square plot)
```

```python
op = rebound.OrbitPlot(sim, orbit_style="solid", lw=1, color= True)
# returns the orbit in a solid line / path
```

```python
sim = rebound.Simulation()
sim.add(m=1.)       # Star A
# it is the primary star at center (represented by the star symbol)

sim.add(m=1., a=1.) # Star B
# Second star orbiting Star A with semi-major axis = 1.
# it forms the binary star system
# it is represented by the orbit in color GOLD

sim.add(a=2.)       # Circumbinary planet ABb
# Planet orbiting both stars together.
# Its orbit encloses the entire binary -> called a circumbinary orbit.
# it is represented by the orbit in color LIME-GREEN

sim.add(a=0.2, primary=sim.particles[1]) # Planet Bb, orbiting star B 
# it is represented by the orbit in color ROYALBLUE
sim.move_to_com()
fig = rebound.OrbitPlot(sim, color=["gold", "royalblue", "limegreen"])
```

```python
ob1 = rebound.OrbitPlot(sim, particles = [1,2], color=True)
# Plots only particles 1 and 2.
# These are drawn relative to the default primary (particle 0).
# Uses automatic coloring.
# Creates a new figure and axes.

# Star B orbiting Star A
# Circumbinary planet orbiting the binary

ob2 = rebound.OrbitPlot(sim, 
particles = [3], 
primary=1, 
show_primary=False, 
fig=ob1.fig, ax = ob1.ax)

# Plots only particle 3
# Sets its reference body to particle 1 (Star B)
# Hides the primary marker
# Draws on the same figure and axes as ob1
```

```python
#Kepler-Orry Orbit Style

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_aspect("equal") # this is needed if you don't want to get stretched orbits
ax.axis('off')  # no axis ticks
for x in range(5):
    for y in range(5):
        sim = rebound.Simulation()
        sim.add(m=1)
        for i in range(3):
            sim.add(a=3+2.2*i, f="uniform")
        sim.move_to_com()
        rebound.OrbitPlot(sim, fig=fig, ax=ax, origin = [-x*20., -y*20.], color=True)
ax.set_xlim([-10,90])
ax.set_ylim([-10,90])

# 25 independent planetary systems
# Each slightly different (random phase)
# Arranged in a clean square grid
# With no axes shown
# This produces a Kepler-style tiled orbit artwork, 
# not a physical simulation of interacting systems.

# useful for
# Teaching Keplerian motion
# Showing orbital symmetry
# Explaining phase differences
```
