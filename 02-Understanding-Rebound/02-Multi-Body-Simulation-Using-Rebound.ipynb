```python
#importing the rebound module
import rebound
import numpy as np

#creating a simulation object
sim = rebound.Simulation()

sim.units = ('AU', 'yr', 'Msun')
```

```python
# New simulation
sim = rebound.Simulation()
sim.units = ('AU', 'yr', 'Msun')

# Central body (acts as origin)
sim.add(m=1.0)

# Orbiting body 1 (inner circle)
sim.add(m=1e-3, a=1.0, e=0.0)

# Orbiting body 2 (outer circle)
sim.add(m=1e-3, a=1.8, e=0.0)

sim.move_to_com()

# Plot in 2D orientation
rebound.OrbitPlot(sim, color = True, figsize=(4,4));
```

```python
# Plot in x, y and z orientation
rebound.OrbitPlotSet(sim, color = True, figsize=(4,4));
```

```python
# New simulation named sim2
sim2 = rebound.Simulation()
sim2.units = ('AU', 'yr', 'Msun')

# Central body (origin)
sim2.add(m=1.0)

# Body 1 → circular orbit
sim2.add(m=1e-3, a=1.0, e=0.0)

# Body 2 → eccentric orbit
sim2.add(m=1e-3, a=1.8, e=0.5)

sim2.move_to_com()

# Plot 2D
rebound.OrbitPlot(sim2, color=True, figsize=(4,4));
```

```python
# Plot in x, y and z axis
rebound.OrbitPlotSet(sim2, color=True, figsize=(4,4));
```

```python
# New simulation
sim2 = rebound.Simulation()
sim2.units = ('AU', 'yr', 'Msun')

# Central body
sim2.add(m=1.0)

# +45 degree inclined circular orbit
sim2.add(m=1e-3, a=1.0, e=0.0, inc=np.pi/4)

# -45 degree inclined circular orbit
sim2.add(m=1e-3, a=1.8, e=0.0, inc=-np.pi/4)

sim2.move_to_com()

# Plot in 2D plane
rebound.OrbitPlot(sim2, color=True, figsize=(4,4));
```

```python
# Plot in x, y and z orientation
rebound.OrbitPlotSet(sim2, color=True, figsize=(4,4));
```

```python
# New simulation
sim4 = rebound.Simulation()
sim4.units = ('AU', 'yr', 'Msun')

# Central body
sim4.add(m=1.0)

# Orbiting bodies (circular, increasing radius, increasing inclination)
sim4.add(m=1e-3, a=1.0, e=0.0, inc=np.deg2rad(10))
sim4.add(m=1e-3, a=1.5, e=0.0, inc=np.deg2rad(20))
sim4.add(m=1e-3, a=2.0, e=0.0, inc=np.deg2rad(30))
sim4.add(m=1e-3, a=2.5, e=0.0, inc=np.deg2rad(40))

sim4.move_to_com()

# plot in 2d orientation
rebound.OrbitPlot(sim4, color=True, figsize=(4,4));
```

```python
# plot in x, y and z orientation
rebound.OrbitPlotSet(sim4, color=True, figsize=(4,4));
```
