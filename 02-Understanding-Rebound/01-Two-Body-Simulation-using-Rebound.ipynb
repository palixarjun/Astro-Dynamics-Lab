```python
#importing the rebound module
import rebound

#creating a simulation object
sim = rebound.Simulation()
```

<img src="../Decorative-Elements/Section-02/ADL-S02-Image-03.png">

```python
#Setting the units for the system
#Distance -> AU (Astronomial units)
#Time -> yr (Years)
#Mass -> Solar Mass (Msun)
sim.units = ('AU', 'yr', 'Msun')
```

```python
# Add central body at origin
sim.add(m=1.0)   # Body A

# Add orbiting body (circular orbit)
sim.add(m=1e-3, a=1.0)  # Body B

sim.move_to_com()
```

```python
# Plot using REBOUND's built-in function
rebound.OrbitPlot(sim, color= True, figsize=(4,4));

# python do not support colons, or ignore them
# but it is used in this instance to suppress a return value 
# <rebound.plotting.OrbitPlot at 0x268baa12710> 
# adding colon suppresses the return value.
```

```python
# Plot using REBOUND's built-in function
rebound.OrbitPlotSet(sim, color= True, figsize=(4,4));
```

```python
# New simulation
sim2 = rebound.Simulation()
sim2.units = ('AU', 'yr', 'Msun')

# Central body
sim2.add(m=1.0)

# Orbiting body with eccentricity
sim2.add(m=1e-3, a=1.0, e=0.5)   # e = 0.5 → elliptical orbit

sim2.move_to_com()

# Plot orbit
rebound.OrbitPlot(sim2, color=True, figsize=(4,4));
```

```python
# Plot orbit, using x, y and z axis
rebound.OrbitPlotSet(sim2, color=True, figsize=(4,4));
```

```python
# New simulation
sim3 = rebound.Simulation()
sim3.units = ('AU', 'yr', 'Msun')

# Central body
sim3.add(m=1.0)

# Orbiting body with inclination
sim3.add(m=1e-3, a=1.0, e=0.3, inc=0.5)  
# inc is in radians (0.5 rad ≈ 28.6 degrees)

sim3.move_to_com()

# Orbit Plot
rebound.OrbitPlot(sim3, color=True, figsize=(4,4));
```

```python
# Orbit Plot, using x, y and z axis
rebound.OrbitPlotSet(sim3, color=True, figsize=(4,4));
```

```python
# New simulation
sim4 = rebound.Simulation()
sim4.units = ('AU', 'yr', 'Msun')

# Central body
sim4.add(m=1.0)

# Orbiting body with omega
sim4.add(m=1e-3, a=1.0, e=0.4, inc=0.5, omega=1.0)
# omega is in radians

sim4.move_to_com()

# Orbit Plot
rebound.OrbitPlot(sim4, color=True, figsize=(4,4));
```

```python
# Orbit Plot, using x, y and z axis
rebound.OrbitPlotSet(sim3, color=True, figsize=(4,4));
```
