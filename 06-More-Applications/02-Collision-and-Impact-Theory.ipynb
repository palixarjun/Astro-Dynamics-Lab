```python
import rebound
import numpy as np
import matplotlib.pyplot as plt
```

```python
sim = rebound.Simulation()
sim.units = ('AU', 'yr', 'Msun')

# Two bodies on collision path
sim.add(m=1e-3, x=-1, vx=1)   # Body A
sim.add(m=1e-3, x=1, vx=-1)   # Body B

sim.move_to_com()
```

```python
sim.collision = "direct"
sim.collision_resolve = "merge"
```

```python
times = np.linspace(0, 2, 200)

x1, y1 = [], []
x2, y2 = [], []

for t in times:
    sim.integrate(t)
    
    if len(sim.particles) > 1:
        x1.append(sim.particles[0].x)
        y1.append(sim.particles[0].y)
        x2.append(sim.particles[1].x)
        y2.append(sim.particles[1].y)
    else:
        break
```

```python
plt.plot(x1, y1, label="Body A")
plt.plot(x2, y2, label="Body B")

plt.scatter(x1[-1], y1[-1], label="Collision Point")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Collision Trajectory")
plt.legend()
plt.show()
```

```python
plt.plot(times[:len(x1)], x1, label="Body A")
plt.plot(times[:len(x2)], x2, label="Body B")

plt.xlabel("Time")
plt.ylabel("Position (x)")
plt.title("Position vs Time")
plt.legend()
plt.show()
```

```python
plt.plot(times[:len(x1)], x1, label="Body A")
plt.plot(times[:len(x2)], x2, label="Body B")

plt.xlabel("Time")
plt.ylabel("Position (x)")
plt.title("Position vs Time")
plt.legend()
plt.show()
```

```python
if len(sim.particles) == 1:
    print("Collision occurred: Bodies merged")
else:
    print("No collision")
```

```python
sim = rebound.Simulation()
sim.units = ('AU', 'yr', 'Msun')

sim.add(m=1e-3, x=-1, vx=1)
sim.add(m=1e-3, x=1, vx=-1)

sim.move_to_com()
sim.collision = "direct"
sim.collision_resolve = "merge"

times = np.linspace(0, 2, 200)

v1, v2 = [], []

for t in times:
    sim.integrate(t)
    
    if len(sim.particles) > 1:
        v1.append(sim.particles[0].vx)
        v2.append(sim.particles[1].vx)
    else:
        break

plt.plot(times[:len(v1)], v1, label="Body A")
plt.plot(times[:len(v2)], v2, label="Body B")

plt.xlabel("Time")
plt.ylabel("Velocity")
plt.title("Velocity vs Time")
plt.legend()
plt.show()
```

```python
plt.plot(x1[:len(v1)], v1, label="Body A")
plt.plot(x2[:len(v2)], v2, label="Body B")

plt.xlabel("Position (x)")
plt.ylabel("Velocity")
plt.title("Phase Space Plot")
plt.legend()
plt.show()
```
