```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Planet: Mercury

# Average distance from Sun: 0.39 AU
# Orbital period around Sun: 0.24 Earth years
# Orbital period in days: 88 days
# Total number of moons: 0

# 0.24 Earth years is equivalent to 88 days.
# Mercury completes its revolution around the Sun in days, not in full Earth years.

# Mercury has no natural satellites.
# It is the smallest planet in the Solar System.
# It has the shortest orbital period among all eight planets.
```

```python
# Mercury orbital radius in AU
mercury_radius = 0.39  

theta = np.linspace(0, 2*np.pi, 400)

# Orbit coordinates
x_orbit = mercury_radius * np.cos(theta)
y_orbit = mercury_radius * np.sin(theta)

plt.figure(figsize=(6,6))

# Sun
plt.scatter(0, 0, s=300, color='gold', label='Sun')

# Orbit path
plt.plot(x_orbit, y_orbit, color='gray', linestyle='--', label='Mercury Orbit')

# Mercury position
plt.scatter(mercury_radius, 0, s=80, color='gray', label='Mercury')

plt.gca().set_aspect('equal', 'box')
plt.title("2D Representation: Sun and Mercury")
plt.xlabel("Distance (AU)")
plt.ylabel("Distance (AU)")

# Legend fixed at upper right
plt.legend(loc='upper right')

plt.grid(True)
plt.show()
```
