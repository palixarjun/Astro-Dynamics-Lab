```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
```

```python
# Planet: Earth

# Average distance from Sun: 1.0 AU
# Orbital period around Sun: 1 Earth year
# Orbital period in days: 365.25 days
# Total number of moons: 1

# One Earth year represents the full orbital revolution around the Sun.
# 1 year is equivalent to 365.25 days.

# Moon: Earth's Moon
# Mean orbital distance from Earth: 384,400 km
# Orbital period (sidereal): 27.3 days
# Orbital period (synodic / phase cycle): 29.5 days

# The Moon revolves around Earth in days, not years.
# 27.3 days is much shorter than Earth's 365.25-day revolution around the Sun.
```

```python
# Earth orbital radius in AU
earth_radius = 1.0  

theta = np.linspace(0, 2*np.pi, 400)

# Orbit coordinates
x_orbit = earth_radius * np.cos(theta)
y_orbit = earth_radius * np.sin(theta)

plt.figure(figsize=(6,6))

# Sun
plt.scatter(0, 0, s=300, color='gold', label='Sun')

# Orbit path
plt.plot(x_orbit, y_orbit, color='blue', linestyle='--', label='Earth Orbit')

# Earth position (example at angle = 0)
plt.scatter(earth_radius, 0, s=100, color='blue', label='Earth')

plt.gca().set_aspect('equal', 'box')
plt.title("2D Representation: Sun and Earth")
plt.xlabel("Distance (AU)")
plt.ylabel("Distance (AU)")

plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```

```python
# Moon orbital radius around Earth (in AU)
moon_radius = 0.00257  

theta = np.linspace(0, 2*np.pi, 400)

# Orbit coordinates
x_orbit = moon_radius * np.cos(theta)
y_orbit = moon_radius * np.sin(theta)

plt.figure(figsize=(6,6))

# Earth at center
plt.scatter(0, 0, s=400, color='blue', label='Earth')

# Moon orbit path
plt.plot(x_orbit, y_orbit, color='gray', linestyle='--', label='Moon Orbit')

# Moon position (example at angle = 0)
plt.scatter(moon_radius, 0, s=80, color='gray', label='Moon')

plt.gca().set_aspect('equal', 'box')
plt.title("2D Representation: Earth and Moon")
plt.xlabel("Distance (AU)")
plt.ylabel("Distance (AU)")

plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```

```python
# (Scaled radius, inclination in degrees)
moon_radius = 2.0
moon_inclination = 5.1  # degrees

theta = np.linspace(0, 2*np.pi, 600)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Earth at origin
ax.scatter(0, 0, 0, s=200, color='blue', label='Earth')

inc = np.radians(moon_inclination)

# Base circular orbit
x = moon_radius * np.cos(theta)
y = moon_radius * np.sin(theta)
z = np.zeros_like(theta)

# Apply inclination (rotate about x-axis)
y_rot = y * np.cos(inc)
z_rot = y * np.sin(inc)

# Plot orbit
ax.plot(x, y_rot, z_rot, color='gray')

# Place Moon at 60° position so tilt is visible
angle = np.pi / 3
moon_x = moon_radius * np.cos(angle)
moon_y = moon_radius * np.sin(angle) * np.cos(inc)
moon_z = moon_radius * np.sin(angle) * np.sin(inc)

ax.scatter(moon_x, moon_y, moon_z, s=50, color='lightgray', label='Moon')

ax.set_title("Static 3D Orientation: Earth and Moon")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

ax.set_box_aspect([1,1,1])
ax.view_init(elev=20, azim=40)

ax.legend(loc="upper right")

plt.show()
```

```python
# Parameters
moon_radius = 2.0
moon_inclination = 5.1  # degrees
theta = np.linspace(0, 2*np.pi, 600)
inc = np.radians(moon_inclination)

# Base circular orbit
x = moon_radius * np.cos(theta)
y_base = moon_radius * np.sin(theta)
z_base = np.zeros_like(theta)

# Apply inclination (rotate about x-axis)
y_rot = y_base * np.cos(inc)
z_rot = y_base * np.sin(inc)

# Moon position at 60° for visibility
angle = np.pi / 3
moon_x = moon_radius * np.cos(angle)
moon_y = moon_radius * np.sin(angle) * np.cos(inc)
moon_z = moon_radius * np.sin(angle) * np.sin(inc)

# Plotly 3D Figure
fig = go.Figure()

# Earth at origin
fig.add_trace(go.Scatter3d(
    x=[0], y=[0], z=[0],
    mode='markers',
    marker=dict(size=15, color='blue'),
    name='Earth'
))

# Moon orbit
fig.add_trace(go.Scatter3d(
    x=x, y=y_rot, z=z_rot,
    mode='lines',
    line=dict(color='gray', width=2),
    name='Moon Orbit'
))

# Moon
fig.add_trace(go.Scatter3d(
    x=[moon_x], y=[moon_y], z=[moon_z],
    mode='markers',
    marker=dict(size=6, color='lightgray'),
    name='Moon'
))

# Layout
fig.update_layout(
    title="Static 3D Orientation: Earth and Moon",
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        aspectmode='data'
    ),
    width=800,
    height=800,
    showlegend=True
)

fig.show()

# Output is not visible in github webpage, refer to Folder/Output-Plotly
```
