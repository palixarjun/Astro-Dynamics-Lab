```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
```

```python
# Planet: Uranus

# Average distance from Sun: 19.2 AU
# Orbital period around Sun: 84 Earth years
# Orbital period in days: 30,687 days
# Total number of confirmed moons: 27

# 84 Earth years is equivalent to approximately 30,687 days.
# Uranus takes much longer than Saturn and Jupiter to complete one revolution around the Sun.

# Rotation period (one spin on its axis): 17.2 hours
# Uranus rotates relatively quickly compared to its very long orbital period.

# Uranus has an axial tilt of about 98 degrees.
# It rotates on its side compared to other planets.
# It has a faint ring system.

# Major Moons:

# Titania
# Orbital period: 8.71 days
# Largest moon of Uranus

# Oberon
# Orbital period: 13.46 days

# Umbriel
# Orbital period: 4.14 days

# Ariel
# Orbital period: 2.52 days

# Miranda
# Orbital period: 1.41 days

# Miranda revolves fastest among the major moons.
# Oberon has the longest orbital period among the listed major moons.
```

```python
uranus_radius = 19.2  # AU

theta = np.linspace(0, 2*np.pi, 400)

x_orbit = uranus_radius * np.cos(theta)
y_orbit = uranus_radius * np.sin(theta)

plt.figure(figsize=(7,7))

# Sun
plt.scatter(0, 0, s=400, color='gold', label='Sun')

# Uranus orbit
plt.plot(x_orbit, y_orbit, linestyle='--', color='black', label='Uranus Orbit')

# Uranus position
plt.scatter(uranus_radius, 0, s=180, color='cyan', label='Uranus')

plt.gca().set_aspect('equal', 'box')
plt.title("2D Representation: Sun and Uranus")
plt.xlabel("Distance (AU)")
plt.ylabel("Distance (AU)")
plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```

```python
# Scaled orbital radii for Uranus moons (visual purpose)
moon_radii = {
    "Miranda": 1.0,
    "Ariel": 1.6,
    "Umbriel": 2.2,
    "Titania": 3.0,
    "Oberon": 3.6
}

theta = np.linspace(0, 2*np.pi, 400)

plt.figure(figsize=(7,7))

# Uranus at center
plt.scatter(0, 0, s=500, color='cyan', label='Uranus')

# Draw orbits and moon positions
for moon, radius in moon_radii.items():
    x_orbit = radius * np.cos(theta)
    y_orbit = radius * np.sin(theta)

    plt.plot(x_orbit, y_orbit, linestyle='--')
    plt.scatter(radius, 0, s=90, label=moon)

plt.gca().set_aspect('equal', 'box')
plt.title("2D Representation: Uranus and Major Moons")
plt.xlabel("Scaled Distance")
plt.ylabel("Scaled Distance")
plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Scaled orbital radii and inclinations (visual purpose)
moon_data = {
    "Miranda": (1.0, 4.3),
    "Ariel": (1.6, 0.0),
    "Umbriel": (2.2, 0.0),
    "Titania": (3.0, 0.0),
    "Oberon": (3.6, 0.0)
}

theta = np.linspace(0, 2*np.pi, 400)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Uranus at origin
ax.scatter(0, 0, 0, s=200, color='cyan', label='Uranus')

for moon, (radius, inclination_deg) in moon_data.items():
    
    inc = np.radians(inclination_deg)
    
    # Base circular orbit in x-y plane
    x = radius * np.cos(theta)
    y = radius * np.sin(theta)
    z = np.zeros_like(theta)
    
    # Rotate orbit around x-axis to create inclination
    y_rot = y * np.cos(inc)
    z_rot = y * np.sin(inc)
    
    # Plot orbit
    ax.plot(x, y_rot, z_rot, label=f'{moon} Orbit')
    
    # Plot moon at one position (45° for visibility)
    angle = np.pi / 4
    moon_x = radius * np.cos(angle)
    moon_y = radius * np.sin(angle) * np.cos(inc)
    moon_z = radius * np.sin(angle) * np.sin(inc)
    
    ax.scatter(moon_x, moon_y, moon_z, s=80, label=moon)

ax.set_title("3D Static Orientation: Uranus and Major Moons")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

ax.legend(loc='upper right')

ax.set_box_aspect([1,1,1])

plt.show()
```

```python
# Moon data: (scaled radius, inclination in degrees)
moon_data = {
    "Miranda": (1.0, 4.3),
    "Ariel": (1.6, 0.0),
    "Umbriel": (2.2, 0.0),
    "Titania": (3.0, 0.0),
    "Oberon": (3.6, 0.0)
}

theta = np.linspace(0, 2*np.pi, 400)

fig = go.Figure()

# Uranus at center
fig.add_trace(go.Scatter3d(
    x=[0], y=[0], z=[0],
    mode='markers',
    marker=dict(size=20, color='cyan'),
    name='Uranus'
))

for moon, (radius, inc_deg) in moon_data.items():
    inc = np.radians(inc_deg)
    
    # Base orbit in xy-plane
    x = radius * np.cos(theta)
    y_base = radius * np.sin(theta)
    z_base = np.zeros_like(theta)
    
    # Inclined orbit about x-axis
    y_rot = y_base * np.cos(inc)
    z_rot = y_base * np.sin(inc)
    
    # Orbit line
    fig.add_trace(go.Scatter3d(
        x=x, y=y_rot, z=z_rot,
        mode='lines',
        line=dict(width=2),
        name=f'{moon} Orbit'
    ))
    
    # Moon position at 45° along orbit for visibility
    angle = np.pi / 4
    moon_x = radius * np.cos(angle)
    moon_y = radius * np.sin(angle) * np.cos(inc)
    moon_z = radius * np.sin(angle) * np.sin(inc)
    
    fig.add_trace(go.Scatter3d(
        x=[moon_x], y=[moon_y], z=[moon_z],
        mode='markers',
        marker=dict(size=6),
        name=moon
    ))

# Layout
fig.update_layout(
    title="3D Static Orientation: Uranus and Major Moons",
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        aspectmode='data'
    ),
    width=800,
    height=800,
    showlegend=True
)

fig.show()

# Output is not visible in github webpage, refer to Folder/Output-Plotly
```
