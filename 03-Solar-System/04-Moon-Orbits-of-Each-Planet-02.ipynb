```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Planet: Venus

# Average distance from Sun: 0.72 AU
# Orbital period around Sun: 0.615 Earth years
# Orbital period in days: 225 days
# Total number of moons: 0

# 0.615 Earth years is equivalent to 225 days.
# Venus completes its revolution around the Sun in less than one Earth year.

# Rotation period (one spin on its axis): 243 Earth days
# Venus rotates more slowly than it orbits the Sun.

# The rotation period (243 days) is longer than the orbital period (225 days).
# Venus completes one orbit around the Sun before finishing one full rotation.

# Venus rotates in the opposite direction (retrograde rotation).
# Venus has no natural satellites.
```

```python
# Venus orbital radius in AU
venus_radius = 0.72  

theta = np.linspace(0, 2*np.pi, 400)

# Orbit coordinates
x_orbit = venus_radius * np.cos(theta)
y_orbit = venus_radius * np.sin(theta)

plt.figure(figsize=(6,6))

# Sun
plt.scatter(0, 0, s=300, color='gold', label='Sun')

# Orbit path
plt.plot(x_orbit, y_orbit, color='orange', linestyle='--', label='Venus Orbit')

# Venus position (example at angle = 0)
plt.scatter(venus_radius, 0, s=90, color='orange', label='Venus')

plt.gca().set_aspect('equal', 'box')
plt.title("2D Representation: Sun and Venus")
plt.xlabel("Distance (AU)")
plt.ylabel("Distance (AU)")

plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```
