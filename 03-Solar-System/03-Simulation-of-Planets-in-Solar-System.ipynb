```python
import rebound
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import plotly.graph_objects as go
import numpy as np
from datetime import datetime
```

```python
def inner_solar_system(date="2026-03-04"):
    
    sim = rebound.Simulation()
    sim.units = ('AU', 'yr', 'Msun')
    
    # Add bodies
    sim.add("Sun", date=date)
    sim.add("Mercury", date=date)
    sim.add("Venus", date=date)
    sim.add("Earth", date=date)
    sim.add("Mars", date=date)
    
    sim.move_to_com()
    
    # Create plot
    fig = rebound.OrbitPlot(
        sim,
        figsize=(5,5),
        color=True,
        periastron=False
    )
    
    # Custom legend
    legend_elements = [
        Line2D([0], [0], marker='o', color='w', label='Sun',
               markerfacecolor='yellow', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Mercury',
               markerfacecolor='gray', markersize=6),
        Line2D([0], [0], marker='o', color='w', label='Venus',
               markerfacecolor='orange', markersize=6),
        Line2D([0], [0], marker='o', color='w', label='Earth',
               markerfacecolor='blue', markersize=6),
        Line2D([0], [0], marker='o', color='w', label='Mars',
               markerfacecolor='red', markersize=6),
    ]
    
    plt.legend(handles=legend_elements, loc="upper right")
    plt.title(f"Inner Solar System on {date}")
    plt.show()

# Run
inner_solar_system("2026-03-04")
```

```python
def outer_solar_system(date="2026-03-04"):
    
    sim2 = rebound.Simulation()
    sim2.units = ('AU', 'yr', 'Msun')
    
    # Add Sun + 4 planets
    sim2.add("Sun", date=date)
    sim2.add("Jupiter", date=date)
    sim2.add("Saturn", date=date)
    sim2.add("Uranus", date=date)
    sim2.add("Neptune", date=date)
    
    sim2.move_to_com()
    
    # Plot
    rebound.OrbitPlot(
        sim2,
        figsize=(5,5),
        color=True,
        periastron=False
    )
    
    # Legend
    legend_elements = [
        Line2D([0], [0], marker='o', color='w', label='Sun',
               markerfacecolor='yellow', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Jupiter',
               markerfacecolor='brown', markersize=8),
        Line2D([0], [0], marker='o', color='w', label='Saturn',
               markerfacecolor='gold', markersize=8),
        Line2D([0], [0], marker='o', color='w', label='Uranus',
               markerfacecolor='lightblue', markersize=7),
        Line2D([0], [0], marker='o', color='w', label='Neptune',
               markerfacecolor='darkblue', markersize=7),
    ]
    
    plt.legend(handles=legend_elements, loc="upper right", fontsize=8)
    plt.title(f"Solar System (2D) on {date}")
    plt.show()


# Run
outer_solar_system("2026-03-04")
```

```python
def full_solar_system(date="2026-03-04"):
    
    sim3 = rebound.Simulation()
    sim3.units = ('AU', 'yr', 'Msun')
    
    # Add Sun + 8 planets
    sim3.add("Sun", date=date)
    sim3.add("Mercury", date=date)
    sim3.add("Venus", date=date)
    sim3.add("Earth", date=date)
    sim3.add("Mars", date=date)
    sim3.add("Jupiter", date=date)
    sim3.add("Saturn", date=date)
    sim3.add("Uranus", date=date)
    sim3.add("Neptune", date=date)
    
    sim3.move_to_com()
    
    # Plot
    rebound.OrbitPlot(
        sim3,
        figsize=(30,30),
        color=True,
        periastron=False
    )
    
    # Legend
    legend_elements = [
        Line2D([0], [0], marker='o', color='w', label='Sun',
               markerfacecolor='yellow', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Mercury',
               markerfacecolor='gray', markersize=6),
        Line2D([0], [0], marker='o', color='w', label='Venus',
               markerfacecolor='orange', markersize=6),
        Line2D([0], [0], marker='o', color='w', label='Earth',
               markerfacecolor='blue', markersize=6),
        Line2D([0], [0], marker='o', color='w', label='Mars',
               markerfacecolor='red', markersize=6),
        Line2D([0], [0], marker='o', color='w', label='Jupiter',
               markerfacecolor='brown', markersize=8),
        Line2D([0], [0], marker='o', color='w', label='Saturn',
               markerfacecolor='gold', markersize=8),
        Line2D([0], [0], marker='o', color='w', label='Uranus',
               markerfacecolor='lightblue', markersize=7),
        Line2D([0], [0], marker='o', color='w', label='Neptune',
               markerfacecolor='darkblue', markersize=7),
    ]
    
    plt.legend(handles=legend_elements, loc="upper right", fontsize=8)
    plt.title(f"Solar System (2D) on {date}")
    plt.show()


# Run
full_solar_system("2026-03-04")
```

```python
planet_data = {
    "Mercury": (0.39, 7.0, "darkgray"),
    "Venus": (0.72, 3.39, "orange"),
    "Earth": (1.00, 0.0, "blue"),
    "Mars": (1.52, 1.85, "red"),
    "Jupiter": (5.20, 1.30, "brown"),
    "Saturn": (9.58, 2.49, "goldenrod"),
    "Uranus": (19.2, 0.77, "lightblue"),
    "Neptune": (30.07, 1.77, "darkblue")
}

theta = np.linspace(0, 2*np.pi, 100)

fig = plt.figure(figsize=(30, 30))
ax = fig.add_subplot(111, projection='3d')

# Sun
ax.scatter(0, 0, 0, s=200, color='gold', label='Sun')

for planet, (radius, inc_deg, color) in planet_data.items():

    inc = np.radians(inc_deg)

    x = radius * np.cos(theta)
    y = radius * np.sin(theta)
    z = np.zeros_like(theta)

    y_rot = y * np.cos(inc)
    z_rot = y * np.sin(inc)

    ax.plot(x, y_rot, z_rot, color=color)

    angle = np.pi / 4
    px = radius * np.cos(angle)
    py = radius * np.sin(angle) * np.cos(inc)
    pz = radius * np.sin(angle) * np.sin(inc)

    ax.scatter(px, py, pz, s=70, color=color, label=planet)

ax.set_title("3D Static Solar System with Orbital Inclinations", fontsize=24)
ax.set_xlabel("X (AU)", fontsize=18)
ax.set_ylabel("Y (AU)", fontsize=18)
ax.set_zlabel("Z (AU)", fontsize=18)

ax.set_box_aspect([1,1,0.4])
ax.view_init(elev=20, azim=35)

ax.legend(loc="upper left", fontsize=14)

plt.show()
```

```python
# ------------------------
# Solar System Data with Inclinations (degrees)
# ------------------------
planets = [
    {"name":"Mercury", "radius":0.39, "period":0.241, "color":"silver", "inc": 7.0},
    {"name":"Venus", "radius":0.72, "period":0.615, "color":"orange", "inc": 3.39},
    {"name":"Earth", "radius":1.0, "period":1.0, "color":"deepskyblue", "inc": 0.0},
    {"name":"Mars", "radius":1.52, "period":1.88, "color":"red", "inc": 1.85},
    {"name":"Jupiter", "radius":5.2, "period":11.86, "color":"darkorange", "inc": 1.30},
    {"name":"Saturn", "radius":9.58, "period":29.46, "color":"gold", "inc": 2.49},
    {"name":"Uranus", "radius":19.2, "period":84.01, "color":"cyan", "inc": 0.77},
    {"name":"Neptune", "radius":30.05, "period":164.8, "color":"royalblue", "inc": 1.77}
]
```

```python
# Time reference
epoch = datetime(2000,1,1)
now = datetime.now()
days_since_epoch = (now - epoch).days
years_since_epoch = days_since_epoch / 365.25

def orbital_angle(years_since_epoch, period):
    fraction = (years_since_epoch / period) % 1
    return 2 * np.pi * fraction
```

```python
# Plot
fig = go.Figure()

# Sun
fig.add_trace(go.Scatter3d(
    x=[0], y=[0], z=[0],
    mode='markers',
    marker=dict(size=20, color='yellow'),
    name='Sun'
))

theta = np.linspace(0, 2*np.pi, 400)

# Planets and Orbits with Inclination
for planet in planets:
    
    inc = np.radians(planet["inc"])
    
    # Orbit in xy-plane
    x_orbit = planet["radius"] * np.cos(theta)
    y_base = planet["radius"] * np.sin(theta)
    z_base = np.zeros_like(theta)
    
    # Apply inclination
    y_orbit = y_base * np.cos(inc)
    z_orbit = y_base * np.sin(inc)
    
    fig.add_trace(go.Scatter3d(
        x=x_orbit, y=y_orbit, z=z_orbit,
        mode='lines',
        line=dict(color=planet["color"], width=2),
        showlegend=False
    ))
    
    # Current position
    angle_now = orbital_angle(years_since_epoch, planet["period"])
    px = planet["radius"] * np.cos(angle_now)
    py = planet["radius"] * np.sin(angle_now) * np.cos(inc)
    pz = planet["radius"] * np.sin(angle_now) * np.sin(inc)
    
    fig.add_trace(go.Scatter3d(
        x=[px], y=[py], z=[pz],
        mode='markers',
        marker=dict(size=6, color=planet["color"]),
        name=planet["name"]
    ))
```

```python
# Asteroid Belt with small random inclinations
num_asteroids = 800
r_inner = 2.1
r_outer = 3.3

r = np.random.uniform(r_inner, r_outer, num_asteroids)
angles = np.random.uniform(0, 2*np.pi, num_asteroids)
inc_deg = np.random.uniform(0, 10, num_asteroids)  # small tilt
inc = np.radians(inc_deg)

x_base = r * np.cos(angles)
y_base = r * np.sin(angles)
y_ast = y_base * np.cos(inc)
z_ast = y_base * np.sin(inc)

fig.add_trace(go.Scatter3d(
    x=x_base, y=y_ast, z=z_ast,
    mode='markers',
    marker=dict(size=2, color='lightgray'),
    name='Asteroid Belt'
))

# Layout
fig.update_layout(
    title=f"3D Solar System with Inclinations and Asteroid Belt — {now.strftime('%Y-%m-%d')}",
    scene=dict(
        xaxis_title='AU',
        yaxis_title='AU',
        zaxis_title='AU',
        aspectmode='data',
    ),
    width=1320,
    height=1020,
    showlegend=True
)

fig.show()

# Output is not visible in github webpage, refer to Folder/Output-Plotly
```
